# Youtube transcript wordcloud

## Summary
- donwload script
- data process text
- filter stopwords
- print wordcloud

```python
# Setup dependencies
from youtube_transcript_api import YouTubeTranscriptApi

from wordcloud import WordCloud, get_single_color_func, ImageColorGenerator

import numpy as np
import matplotlib
import matplotlib.pyplot as plt

import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
# downlaod transcript
video_id = 'nSSSqJVTseA' # hOfRN0KihOU nSSSqJVTseA
content = YouTubeTranscriptApi.get_transcript(video_id)
content
```

```python
text_content = [sentence['text'] for sentence in content]
text_string = ''.join(text_content)
text_string = text_string.split()
text_string
```

```python
# filter stop words
filtered_words = [word for word in text_string if word not in stopwords.words('german')]
final_text = ' '.join(filtered_words)
final_text
```

```python
# print wordcloud
wordcloud = WordCloud(max_font_size=40).generate(final_text)
plt.figure(figsize=(24,24))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

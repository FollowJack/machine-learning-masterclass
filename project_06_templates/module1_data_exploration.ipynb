# Distributions

## Distributions of numerical features

```python
df.hist(xrot=-45, figsize=(10,10))
plt.show()
```

## Distributions of categorical features
Many real-world phenomena follow the normal distribution, or approximately do so, so its a very useful distribution for many practical real world scenarios.

```python
df.describe(include=['object'])
```

Plot every object feature

```python
for features in df.dtypes[df.dtypes == 'object'].index:
    sns.countplot(y=features, data=df)
    plt.show()
```

### Segementations
Segmentations are powerful ways to observe relationship between categorical and numerical features

```python
sns.violinplot(y='status', x='satisfaction', data=df)
```

```python
sns.violinplot(y='status', x='last_evaluation',data=df)
```

### See averages (like the hours spending per month)
### Group by status 'left','employed' and get the mean() 

```python
df.groupby('status').mean()
```

### Advanced Segmentation (bivariate segmentations)

If target variable is categorical then helpful to expand segmentation analysis

```python
sns.lmplot(y='satisfaction', x='last_evaluation', data=df[df.status == 'Employed'], hue='status', fit_reg=None)
```

```python
sns.lmplot(y='satisfaction', x='last_evaluation', data=df[df.status == 'Left'], hue='status', fit_reg=None)
```
